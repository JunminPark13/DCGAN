```python
import os
import torch
import torch.nn as nn
from torch.optim import Adam
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
from torch.autograd import Variable
import torchvision.utils as vutils
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
import matplotlib.animation as animation
%matplotlib inline
```

```python
trans = transforms.Compose([
    transforms.Resize((64,64)),
    transforms.ToTensor(),
    transforms.Normalize(mean=(0.5,), std=(0.5,))
])
dset = datasets.ImageFolder(root='./train/', transform=trans)
dataloader = DataLoader(dset, batch_size=8, shuffle=True)
```

```python
def convert_to_imshow_format(image):
    # first convert back to [0,1] range from [-1,1] range
    image = image / 2 + 0.5
    image = image.numpy()
    # convert from CHW to HWC
    # from 3x32x32 to 32x32x3
    return image.transpose(1,2,0)

temp_loader = torch.utils.data.DataLoader(dset, batch_size=30, shuffle=True)

dataiter = iter(temp_loader)
images, labels = dataiter.next()

images = images[:6]
labels = labels[:6]

fig, axes = plt.subplots(1, len(images), figsize=(12,2.5))
for idx, image in enumerate(images):
    axes[idx].imshow(convert_to_imshow_format(image))
    axes[idx].set_xticks([])
    axes[idx].set_yticks([])
```

```python
import imageio

if torch.cuda.is_available(): use_gpu = True
    
leave_log = True

if leave_log:
    result_dir = 'DCGAN_generated_images'
    if not os.path.isdir(result_dir):
        os.mkdir(result_dir)
```

```python
# 트레이닝 사진 채널 갯수
nc = 3

# 랜덤 노이즈 사이즈
nz = 100

# Size of feature maps in generator
ngf = 64

# Size of feature maps in discriminator
ndf = 64

# 사용가능 GPU 갯수
ngpu = 1
```

```python
# Generator network

class Generator(nn.Module):
    def __init__(self):
        super(Generator, self).__init__()
        self.main = nn.Sequential(
            # input is Z
            nn.ConvTranspose2d( nz, ngf * 8, 4, 1, 0, bias=False),
            nn.BatchNorm2d(ngf * 8),
            nn.ReLU(True),
            # (ngf*8) x 4 x 4
            nn.ConvTranspose2d(ngf * 8, ngf * 4, 4, 2, 1, bias=False),
            nn.BatchNorm2d(ngf * 4),
            nn.ReLU(True),
            # (ngf*4) x 8 x 8
            nn.ConvTranspose2d( ngf * 4, ngf * 2, 4, 2, 1, bias=False),
            nn.BatchNorm2d(ngf * 2),
            nn.ReLU(True),
            # (ngf*2) x 16 x 16
            nn.ConvTranspose2d( ngf * 2, ngf, 4, 2, 1, bias=False),
            nn.BatchNorm2d(ngf),
            nn.ReLU(True),
            # (ngf) x 32 x 32
            nn.ConvTranspose2d( ngf, nc, 4, 2, 1, bias=False),
            nn.Tanh()
            # (nc) x 64 x 64
        )

    def forward(self, input):
        return self.main(input)
```

```python
# Discriminator network 생성

class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()

        self.main = nn.Sequential(
            # (nc) x 64 x 64
            nn.Conv2d(nc, ndf, 4, 2, 1, bias=False),
            nn.LeakyReLU(0.2, inplace=True),
            # (ndf) x 32 x 32
            nn.Conv2d(ndf, ndf * 2, 4, 2, 1, bias=False),
            nn.BatchNorm2d(ndf * 2),
            nn.LeakyReLU(0.2, inplace=True),
            # (ndf*2) x 16 x 16
            nn.Conv2d(ndf * 2, ndf * 4, 4, 2, 1, bias=False),
            nn.BatchNorm2d(ndf * 4),
            nn.LeakyReLU(0.2, inplace=True),
            # (ndf*4) x 8 x 8
            nn.Conv2d(ndf * 4, ndf * 8, 4, 2, 1, bias=False),
            nn.BatchNorm2d(ndf * 8),
            nn.LeakyReLU(0.2, inplace=True),
            # (ndf*8) x 4 x 4
            nn.Conv2d(ndf * 8, 1, 4, 1, 0, bias=False),
            nn.Sigmoid()
        )

    def forward(self, input):
        return self.main(input)
```

```python
### Generator, Discriminator 객체 만들기

G = Generator()
D = Discriminator()

if torch.cuda.is_available():
    G.cuda()
    D.cuda()
```

```python
# Binary Cross Entropy loss

criterion = nn.BCELoss()

G_optimizer = Adam(G.parameters(), lr=0.0005, betas=(0.5, 0.999))
D_optimizer = Adam(D.parameters(), lr=0.0001, betas=(0.5, 0.999))
```

```python
# random vector 정의

z_fixed = Variable(torch.randn(64, 100, 1, 1), volatile=True)
if torch.cuda.is_available(): z_fixed = z_fixed.cuda()
```

```python
img_list = []

# Training

for epoch in range(100):
    
    if leave_log:
        D_losses = []
        G_losses = []
    
    for real_data, _ in dataloader:
        batch_size = real_data.size(0)
        
        real_data = Variable(real_data)

        ### Discriminator

        # 이미지가 진짜일 때 label 값은 1이고 가짜일 때는 0이다.
        target_real = Variable(torch.ones(batch_size, 1, 1, 1))
        target_fake = Variable(torch.zeros(batch_size, 1, 1, 1))
         
        if torch.cuda.is_available():
            real_data, target_real, target_fake = real_data.cuda(), target_real.cuda(), target_fake.cuda()
            
        D_result_from_real = D(real_data)
        D_loss_real = criterion(D_result_from_real, target_real)

        # random noise z
        z = Variable(torch.randn((batch_size, 100, 1, 1)))
        
        if torch.cuda.is_available():
            z = z.cuda()
            
        fake_data = G(z)
        
        # generator가 만든 가짜 이미지를 discriminator에 넣는다.
        D_result_from_fake = D(fake_data)
        D_loss_fake = criterion(D_result_from_fake, target_fake)
        
        D_loss = D_loss_real + D_loss_fake
        
        D.zero_grad()
        D_loss.backward()
        D_optimizer.step()
        
        if leave_log:
            D_losses.append(D_loss)

        # Generator

        z = Variable(torch.randn((batch_size, 100, 1, 1)))
        
        if torch.cuda.is_available():
            z = z.cuda()
        
        # 가짜 이미지 생성
        fake_data = G(z)
        D_result_from_fake = D(fake_data)
        G_loss = criterion(D_result_from_fake, target_real)
        
        G.zero_grad()
        G_loss.backward()
        G_optimizer.step()
        
        if leave_log:
            G_losses.append(G_loss)
            

    if leave_log:
        true_positive_rate = (D_result_from_real > 0.5).float().mean().item()
        true_negative_rate = (D_result_from_fake < 0.5).float().mean().item()
        base_message = ("Epoch: {epoch:<3d} D Loss: {d_loss:<8.6} G Loss: {g_loss:<8.6} "
                        "True Positive Rate: {tpr:<5.1%} True Negative Rate: {tnr:<5.1%}"
                       )
        message = base_message.format(
                    epoch=epoch,
                    d_loss=sum(D_losses)/len(D_losses),
                    g_loss=sum(G_losses)/len(G_losses),
                    tpr=true_positive_rate,
                    tnr=true_negative_rate
        )
        print(message)
    
    
    with torch.no_grad():
        fake = G(z_fixed).detach().cpu()
        img_list.append(vutils.make_grid(fake, padding=2, normalize=True))


# training된 G와D를 저장
torch.save(G.state_dict(), "dcgan_generator.pth")
torch.save(D.state_dict(), "dcgan_discriminator.pth")
```

```python
# result 확인

fig = plt.figure(figsize=(8,8))
plt.axis("off")

for i in range(len(img_list)):
    plt.imshow(np.transpose(img_list[i], (1,2,0)), animated=True)
    plt.savefig(result_dir + '/epoch{}.png'.format(i+1))
    
```
